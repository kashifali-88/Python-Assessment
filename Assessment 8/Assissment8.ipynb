```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"
```

```python
print(calculate_grade(15))
```

```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
```

```python
print(is_even(10))
print(is_even(15))
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
print(factorial(7))
```
