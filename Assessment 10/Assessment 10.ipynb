** How many people made the purchase during the AM and how many people made the purchase during PM ? **

**(Hint: Check out [value_counts()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) ) **

```python
purchase_counts = ecom['AM or PM'].value_counts()

num_am_purchases = purchase_counts['AM']
num_pm_purchases = purchase_counts['PM']

print("PM:", num_pm_purchases)
print("AM:", num_am_purchases)
```

** Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction? **

** What is the email of the person with the following Credit Card Number: 4926535242672853 **

```python
# Condition 1: Card Provider is 'American Express'
# Condition 2: Purchase Price is greater than 95
amex_high_purchase = ecom[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)]

# Count the number of rows that match
count = len(amex_high_purchase)
print(f"Number of people: {count}")
```

** How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
num_people = ((ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)).sum()

num_people
```

** Hard: How many people have a credit card that expires in 2025? **

```python
ecom['CC Exp Year'] = pd.to_datetime(ecom['CC Exp Date'], format='%m/%y').dt.year

# Count the occurrences where the year is '2025'
num_people = (ecom['CC Exp Year'] == 2025).sum()

num_people
```

** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
ecom['Email Domain'] = ecom['Email'].str.split('@').str[1]

# Count the occurrences of each email domain
top_email_providers = ecom['Email Domain'].value_counts().head(5)

top_email_providers
```
