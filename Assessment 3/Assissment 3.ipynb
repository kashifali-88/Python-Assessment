```python
# Q2: Swap two variables using a third variable

# Taking input from user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

# Swapping using a temporary variable
temp = x
x = y
y = temp

print("After swapping:")
print("x:", x)
print("y:", y)
```

```python
# Q3: Swap two variables without using a third variable

# Taking input from user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

# Swapping without a third variable
x, y = y, x

print("After swapping:")
print("First number:", x)
print("Second number:", y)
```

```python
# Q4: Convert temperature from Celsius to Fahrenheit

# Taking input from user
celsius = float(input("Enter temperature in Celsius: "))

# Formula: F = (9/5) * Celsius + 32
fahrenheit = (9/5) * celsius + 32

print("Temperature in Fahrenheit:", fahrenheit)
```

```python
# Q5: Calculate the sum of three numbers

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

total_sum = num1 + num2 + num3

print("The sum of the three numbers is:", total_sum)
```
