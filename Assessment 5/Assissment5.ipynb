```python
fruits = input("Enter three of your favorite fruits (comma-separated): ")
fruit_list = fruits.split(",")
```

```python
print("Your list of fruits:", fruit_list)
```

```python
print("First fruit:", fruit_list[0])
print("Last fruit:", fruit_list[-1])
print("Second fruit:", fruit_list[1])
print("First two fruits:", fruit_list[:2])
```

```python
fruit_list.append("strawberry")
print("Updated list after append:", fruit_list)
```

```python
fruit_list.remove(fruit_list[1])
print("Updated list after remove:", fruit_list)
```

```python
print("Length of the list is now:", len(fruit_list))
```

```python
letters_list = list(full_name.replace(" ", ""))
print("First 3 letters:", letters_list[:3])
print("Last 3 letters:", letters_list[-3:])
```

```python
letters_list[:3] = [letter.upper() for letter in letters_list[:3]]
letters_list[3:] = [letter.lower() for letter in letters_list[3:]]
modified_name = "".join(letters_list)
print("Modified name:", modified_name)
```

```python
numbers = input("Enter 5 numbers (space-separated): ")
numbers_list = [int(num) for num in numbers.split()]
print("Sum of your numbers:", sum(numbers_list))
print("Average of your numbers:", sum(numbers_list) / len(numbers_list))
```
