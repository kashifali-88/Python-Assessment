```python
wallet = 50
item_price = 15
tax_rate = 0.03
total_cost = item_price + (item_price * tax_rate)
money_left = wallet - total_cost
print(f"After buying the item, you have ${money_left} left.")
```

```python
days_input = input("How many days until your birthday? ")
days = int(days_input)
weeks = days / 7
print(f"That is approximately {weeks:.1f} weeks away!")
```

```python
grade = 85 
if 90 <= grade <= 100:
    print("Letter Grade: A")
elif 80 <= grade <= 89:
    print("Letter Grade: B")
elif 70 <= grade <= 79:
    print("Letter Grade: C")
elif 60 <= grade <= 69:
    print("Letter Grade: D")
else:
    print("Letter Grade: F")
```
